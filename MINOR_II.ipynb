```python
from PIL import Image

# Image load karo (Colab me upload ki ho to path change karna)
img = Image.open('/content/WhatsApp Image 2026-01-03 at 9.57.01 PM.jpeg')

# Size chhota karo (zyada bada hoga to output messy hoga)
width = 60

aspect_ratio = img.height / img.width
height = int(width * aspect_ratio * 0.5)
img = img.resize((width, height))

# Grayscale me convert
img = img.convert('L')

# Pixels nikaalo
pixels = list(img.getdata())

# Threshold (dark = *, light = space)
threshold = 120

# ASCII art print
for i in range(0, len(pixels), width):
    row = pixels[i:i+width]
    line = "".join('*' if pixel < threshold else ' ' for pixel in row)
    print(line)
```
